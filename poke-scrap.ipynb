```python
def pokeprueba(lista):
    lista_total = []
    for nombre in lista:
        url =f'https://bulbapedia.bulbagarden.net/wiki/{nombre}_(Pok%C3%A9mon)'

        respuesta = requests.get(url)
        print(f'pokemon: {nombre}. respuesta: {respuesta.status_code}')
        sopa = BeautifulSoup(respuesta.content, 'html.parser')
        lista_total.append(sopa)
    
    return lista_total
```

```python
def extraer_info(lista):
    lista_total = []
    for nombre in lista:
        url =f'https://bulbapedia.bulbagarden.net/wiki/{nombre}_(Pok%C3%A9mon)'

        respuesta = requests.get(url)
        print(f'pokemon: {nombre}. respuesta: {respuesta.status_code}')
        sopa = BeautifulSoup(respuesta.content, 'html.parser')
        tabla = sopa.find_all('table', {'class':"roundy"})
    
        for elemento in tabla:
            elemento = elemento.text.strip()
            elemento = elemento.replace('\n', ' ').replace('Unknown', '')
            elemento = re.sub(r'\s{2,}', ' ', elemento)

            lista_total.append(elemento.split(' '))
    
    
    return lista_total
```
